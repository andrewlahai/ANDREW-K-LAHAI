# Problem 1] Number of grains of rice on the 100th day

```python
import numpy as np
arr =[1]
total_num_days = 100
for i in range(1,total_num_days):
    arr.append((arr[i-1])*2)
arr=np.array(arr)
num_grains=np.sum(arr)
print("Total number of rice grains = {}".format(num_grains))
```

```python
#Drawing a line graph to represent the change in the number of rice grains you received
import matplotlib.pyplot as plt

num_days = 101
days = range(1, num_days)

sum_arr=[]
for i in range(100):
    sum_arr.append(np.sum(arr[:i+1]))

plt.plot(days, arr, 'g-o', linewidth=3, markersize=7, markeredgewidth=3, markeredgecolor='BLUE')
plt.plot(days, sum_arr, 'b^', linewidth=3, markersize=7);
plt.title('AMOUNT of Rice grains')
plt.xlabel('Days'); 
plt.ylabel('AMOUNT of Rice grains');
plt.legend(['AMOUNT of rice grains you receive on that day', 'The total AMOUNT of rice grains you receive by that day'])
plt.show()
```
