# Problem 1] Creation of random numbers

```python
import numpy as np
mean1 = (-3,0)
cova = [[1,0.8],[0.8,1]]
np.random.seed(0)

random_ndarray = np.random.multivariate_normal(mean1,cova,size=500)

print(random_ndarray.shape)
print(random_ndarray)
```

# Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

x, y = np.random.multivariate_normal(mean1, cova, 700).T
plt.scatter(x, y, color='BLUE', alpha=0.5)
plt.show()
```

# Problem 3] Visualization by histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histogram of x1")

x, y = np.random.multivariate_normal(mean1, cova, 700).T
plt.hist(x, color="BLUE")
plt.xlim(right=5)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histogram of x2")

x, y = np.random.multivariate_normal(mean1, cova, 700).T
plt.hist(y, color="BLUE")
plt.xlim(left=-6)
plt.show()
```

# Problem 4] Addition of data

```python
mean4 = (0,-3)
cova = [[1,0.8],[0.8,1]]
np.random.seed(0)
second_random_ndarray = np.random.multivariate_normal(mean4,cova,size=700)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")

x, y = np.random.multivariate_normal(mean1, cova, 700).T
x2, y2 = np.random.multivariate_normal(mean4, cova, 700).T
classes = [0,1]

plt.scatter(x, y, color="BLUE", alpha=0.5)
plt.scatter(x2, y2, color="YELLOW")
plt.legend(labels=classes)
plt.show()
```

# Problem 5] Data combination

```python
random_ndarray = np.random.multivariate_normal(mean1,cova,size=700)
second_random_ndarray = np.random.multivariate_normal(mean4,cova,size=700)

np.random.seed(0)
combined_ndarray=np.concatenate((random_ndarray, second_random_ndarray))

print(combined_ndarray.shape)
print(combined_ndarray)
```

# Problem 6] Labeling

```python
zeros_ndarray = np.zeros(700)
ones_ndarray = np.ones(700)

first_ndarray=np.insert(random_ndarray,2,zeros_ndarray,axis=1)
second_ndarray=np.insert(second_random_ndarray,2,ones_ndarray,axis=1)
np.random.seed(0)
labeled_ndarray=np.concatenate((first_ndarray, second_ndarray))

print(labeled_ndarray.shape)
print(labeled_ndarray)
```
